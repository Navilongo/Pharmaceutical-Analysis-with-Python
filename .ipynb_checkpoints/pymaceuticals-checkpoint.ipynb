## Observations and Insights 

**Analysis of data:** 
* Judging from the Final Volume per regimen, Ramicane and Capomulin were the most efficient at lowering the size of the tumor in mice.


* The distribution by sex is nearly identical 51% Male vs. 49% Female, which was made to get a more accurate sense of the results of the regimens or medications across either sex, regarless of male/female hormones.


* Lastly, judging by the number of measurements taken with each regimen as compared to the "Placebo" it would be important to note the reason why the other regimens did not have a higher number of measurements; this may be impacted by the passing of the rats as an effect of the cancerous tumor, therefore less measurements were able to be taken. My initial recommendation, based on the studies on these mice, would be to encourage the use of Ramicane or Capomulin in human trials to gauge their effectiveness and continue to understand the effectiveness in reducing tumor volume in patients. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import pylab
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.merge(mouse_metadata, study_results,
                                 how='outer', on='Mouse ID')
```

```python
# Find any duplicate mouse IDs (where Mouse ID and Timepoint repeat) 
# and remove them from the data set. 
clean_mouse_df = mouse_df.drop_duplicates(subset=['Mouse ID', 'Timepoint' ])
```

## Summary Statistics

```python
# Use regimen as index to create a statistical table. 
drug_regimen_df = clean_mouse_df.groupby('Drug Regimen')

# Cast mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen

mean_regimen = round(drug_regimen_df['Tumor Volume (mm3)'].mean(), 2)
median_regimen = round(drug_regimen_df['Tumor Volume (mm3)'].median(), 2)
var_regimen = round(drug_regimen_df['Tumor Volume (mm3)'].var(), 2)
std_regimen = round(drug_regimen_df['Tumor Volume (mm3)'].std(), 2)
sem_regimen = round(drug_regimen_df['Tumor Volume (mm3)'].sem(), 2)
```

```python
# Use statistical data compiled and create a Summary Statistical Table
summary_stats_regimen = pd.DataFrame({
    'Mean': mean_regimen,
    'Median': median_regimen,
    'Variance': var_regimen,
    'Standard Deviation': std_regimen,
    'SEM': sem_regimen
})
summary_stats_regimen.head()
```

## Quartiles, Outliers and Boxplots

```python
# From the non-duplicate df, leave only the rows with the last Tumor Volume value to create statistical data
tumor_mouse_df = clean_mouse_df.drop_duplicates(subset=['Mouse ID'], keep='last')
tumor_mouse_df

```

```python
# Use regimen as index to create a statistical table. 
tumor_regimen_only = tumor_mouse_df[['Drug Regimen', 'Timepoint', 'Tumor Volume (mm3)']]
tumor_regimen_only.head(5)
```

```python
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
#df.loc[df['column_name'] == some_value]
tumor_volume = []


for treatment in treatments:
    top_regimens = tumor_regimen_only.loc[tumor_regimen_only['Drug Regimen'] == treatment, "Tumor Volume (mm3)"]
    tumor_volume.append(top_regimens)

    
    quartiles = top_regimens.quantile([.25, .5, .75])
    lowerq = round(quartiles[0.25], 2)
    upperq = round(quartiles[0.75], 2)
    iqr = round(upperq-lowerq, 2)

    print(f"The lower quartile of the tumor volume in {treatment} is: {lowerq}")
    print(f"The upper quartile of tumor volume is: {upperq}")
    print(f"The interquartile range of tumor volume is: {iqr}")

    lower_limit = round(lowerq - (1.5*iqr), 2)
    upper_limit = round(upperq + (1.5*iqr), 2)
    print(f"Values below {lower_limit} could be outliers.")
    print(f"Values above {upper_limit} could be outliers.")
```

```python

plt.boxplot(tumor_volume)
plt.title('Tumor Volumes per Regimen')
plt.ylabel("Final Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
x_values = mouse_id_group["Weight (g)"]
y_values = mouse_id_group["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
plt.scatter(x_values, y_values)
plt.scatter(mouse_id_group["Weight (g)"], mouse_id_group["Tumor Volume (mm3)"])
plt.plot(mouse_id_group["Weight (g)"],regress_values, "r-")
plt.annotate(line_eq,(18,30), fontsize=15, color="blue")
plt.title("Volume vs. Weight")
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume (mm3)")
```
